```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
print("Shape:", df.shape)
```

```python
print("First 5 rows:\n", df.head())
```

```python
print("\nData Info:")
print(df.info())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#Visualization- Age Histogram
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Visualization - Churn Rate by Geography
plt.figure(figsize=(6, 4))
sns.barplot(x='Geography', y='Exited', hue='Geography', legend=False,
            data=df, estimator=lambda x: sum(x)/len(x),
            errorbar=None, palette='pastel')
plt.title('Churn Rate by Geography')
plt.ylabel('Churn Rate')
plt.xlabel('Geography')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
#Visualization - Balance Box Plot by Churn
plt.figure(figsize=(6, 4))
sns.boxplot(x='Exited', y='Balance', data=df,
            hue='Exited', legend=False, palette='Set2')
plt.title('Balance Distribution by Churn')
plt.xlabel('Exited (0 = Stayed, 1 = Left)')
plt.ylabel('Account Balance')
plt.tight_layout()
plt.show()
```

```python
#Visualization - Heatmap of Correlations
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
#Visualization - Gender vs Churn
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', hue='Exited', data=df, palette='Set1')
plt.title('Churn Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
